```python
# I found a dataset for a University looking into creating a better lead scoring model. This is my approach.
# It is of roughly 10,000 individuals
# There are ~15x more non-conversions than conversions. The data is unbalanced
# Primary KPI: "Lead Grade" == "closed"

# I will build a random forest, a random forest with randomized search, and a gradient boosted model

# The data can be found here: https://www.kaggle.com/rockbottom73/leads-dataset?select=SampleData.csv
```

```python
lead = pd.read_csv("LeadScore.csv", low_memory=False)
lead.dropna(how="all", axis=1, inplace=True)

# Drop Duplicates and nearly-empty columns
lead.drop_duplicates(subset=None, inplace=True)
lead.dropna(thresh=400, axis=1, inplace=True)

lead.head()
```

```python
lead.drop(
    ["Prospect ID", "Lead Number", "First Landing Page Submission Id", "Country"],
    axis=1,
    inplace=True,
)

# Dropping all these columns for not being relevant enough // for being redundant

lead.drop(["Lead Conversion Date"], axis=1, inplace=True)
lead.drop(["Cityold"], axis=1, inplace=True)
lead.drop(["Country New"], axis=1, inplace=True)
lead.drop(["Admission Number"], axis=1, inplace=True)
lead.drop(
    [
        "Last Notable Activity Date",
        "Last Notable Activity",
        "First Landing Page Submission Date",
        "Last Visit Date",
        "Last Activity Date",
    ],
    axis=1,
    inplace=True,
)
```

```python
# Calculated Elapsed Duration between start of funnel and last activity
# IF data were live, we could calculated elapsed time until last contact

lead["Created On"] = pd.to_datetime(lead["Created On"], format="%d-%m-%Y %H:%M").dt.date
lead["Modified On"] = pd.to_datetime(
    lead["Modified On"], format="%d-%m-%Y %H:%M"
).dt.date

elapsed_time = lead["Modified On"] - lead["Created On"]
lead["Elasped Time"] = elapsed_time
lead["Elasped Time"] = (lead["Elasped Time"] / np.timedelta64(1, "D")).astype(int)
lead.drop(["Created On", "Modified On"], axis=1, inplace=True)
```

```python
# Remove All Columns were all fields are the same (just one answer repeated, unique count of 1)

nunique = lead.nunique()
cols_to_drop = nunique[nunique == 1].index
lead.drop(cols_to_drop, axis=1, inplace=True)
```

```python
lead["Notes"] = lead["Notes"].astype(str)
lead["Last Degree"] = lead["Last Degree"].astype(str)
lead["Notes"] = lead["Notes"].str.lower()
lead["Last Degree"] = lead["Last Degree"].str.lower()

# Here is am categorizing string data based on keywords
# Open answers to questions provided by prospects


def checkNotes(x):
    if ("fee" in x) or ("tuition" in x) or ("price" in x) or ("cost" in x):
        x = "fee"
    elif ("mba" in x) or ("management" in x):
        x = "mba"
    elif ("contact" in x) or ("call" in x):
        x = "contact"
    elif ("admission" in x) or ("apply" in x):
        x = "admission"
    elif ("course" in x) or ("details" in x) or ("program" in x) or ("class" in x):
        x = "course"
    else:
        x = 0
    return x


lead["Notes"] = lead["Notes"].apply(lambda x: checkNotes(x))


def checkDegree(x):
    if "mba" in x:
        x = "mba"
    elif "com" in x:
        x = "com"
    elif "tech" in x:
        x = "tech"
    elif "msc" in x:
        x = "msc"
    elif ("b sc" in x) or ("bsc" in x):
        x = "bsc"
    elif "bba" in x:
        x = "bba"
    elif "ba" in x:
        x = "ba"
    elif "be" in x:
        x = "be"
    else:
        x = 0
    return 0


lead["Last Degree"] = lead["Last Degree"].apply(lambda x: checkDegree(x))


def checkState(x):
    if "Maharashtra" in x:
        x = "Maharashtra"
    elif "Gujarat" in x:
        x = "Gujarat"
    elif "Haryana" in x:
        x = "Haryana"
    elif "Karnataka" in x:
        x = "Karnataka"
    elif "Madhya Pradesh" in x:
        x = "Madhya Pradesh"
    else:
        x = 0
    return 0


lead["State"] = lead["State"].apply(lambda x: checkState(str(x)))
```

```python
# What percent of each column is null? Review and drop cols with high # of nulls
round(100 * (lead.isnull().sum() / len(lead.index)), 2)
```

```python
# Looked over columns with high % null and dropping columns

lead.drop(
    ["Source Content", "Lead Grade", "Admission Constraints"], axis=1, inplace=True
)
lead.drop(["Source Campaign", "Related Landing Page Id"], axis=1, inplace=True)

# Replace NaNs
lead = lead.replace("Select", np.nan)
lead = lead.replace(np.nan, 0)
```

```python
# Drop the columns related to the scoring system of the organization's own sales team
lead.drop(
    [
        "Asymmetrique Activity Index",
        "Asymmetrique Activity Score",
        "Asymmetrique Profile Index",
        "Asymmetrique Profile Score",
    ],
    axis=1,
    inplace=True,
)
```

```python
# When Lead Stage == Closed with have a conversion (this is our KPI)

lead["Lead Stage"] = lead["Lead Stage"].apply(lambda x: 1 if "Closed" in str(x) else 0)
```

```python
# Drop these highly unbalanced Bools
lead.drop(
    [
        "Through Recommendations",
        "Newspaper",
        "Newspaper Article",
        "Digital Advertisement",
        "Welearn Forums",
        "Search",
        "Do Not Call",
        "Do Not Email",
    ],
    axis=1,
    inplace=True,
)
```

```python
# Fix Entrance Test;; Almost all answers are either MAT or blank
lead["Entrance Test"] = lead["Entrance Test"].apply(
    lambda x: 1 if "MAT" in str(x) else 0
)
```

```python
# Update data types
lead = lead.convert_dtypes()
for col in lead:
    if lead[col].dtype == "object":
        lead[col] = lead[col].astype(str)

# This code has to be run twice for some reason.... look into this
lead = lead.convert_dtypes()
for col in lead:
    if lead[col].dtype == "object":
        lead[col] = lead[col].astype(str)
```

```python
# Convert Lead Stage (our primary KPI/target) to a boolean

df_converted = lead.loc[lead["Lead Stage"] == 0]
df_notconverted = lead.loc[lead["Lead Stage"] == 1]

# There are 16x more non-conversions than conversions
# This is unbalanaced
# So I will be choosing AUCPR over AUC
# AUCPR is less sensitive to a high percentage of True Negatives
```

```python
# STANDARDIZE DATA -- scaling data for sklearn
lead2 = lead

scaler = StandardScaler()

lead2[
    [
        "TotalVisits",
        "Average Time Per Visit",
        "Page Views Per Visit",
        "Engagement Score",
        "Lead Score",
    ]
] = scaler.fit_transform(
    lead2[
        [
            "TotalVisits",
            "Average Time Per Visit",
            "Page Views Per Visit",
            "Engagement Score",
            "Lead Score",
        ]
    ]
)
```

```python
# Creating Dummy Variables
for col in lead2:
    try:
        if lead2[col].dtypes == "string":
            dummy = pd.get_dummies(lead2[col], drop_first=True)
            lead2 = pd.concat([lead2, dummy], axis=1)
            lead2.drop([col], axis=1, inplace=True)
    except:
        pass
```

```python
# Code copied from: https://stackoverflow.com/questions/29294983/how-to-calculate-correlation-between-all-columns-and-remove-highly-correlated-on

# Create correlation matrx
corr_matrix = lead2.corr().abs()
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
to_drop = [column for column in upper.columns if any(upper[column] > 0.95)]

# Drop cols with a correlation >95%
lead2.drop(to_drop, axis=1, inplace=True)

corr_matrix = lead2.corr().abs()
plt.matshow(lead2.corr().abs())
plt.show()

# Deprecation Warning
sol = (
    corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
    .stack()
    .sort_values(ascending=False)
)
```

```python
# This will show us other highly correlated columns
print(sol)
```

```python
# Dropping this confounding columns which is nearly 1:1 with the KPI
lead2.drop(["Closed by Horizzon"], axis=1, inplace=True)
```

<center><font size="10">MODEL #1: Random Forest</font></center>


```python
model_rf = RandomForestClassifier(
    n_estimators=400,
    class_weight="balanced_subsample",  # We have unbalanced data!
    criterion="gini",
    bootstrap=False,
    max_features="sqrt",
    min_samples_split=2,
    min_samples_leaf=1,
    max_depth=5,
    n_jobs=-1,
)

X = lead2.drop(["Lead Stage"], axis=1)
y = lead2["Lead Stage"]
feature_list = list(X.columns)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5)

rf = model_rf.fit(X_train, y_train.astype("int"))

AUC = roc_auc_score(y_test.astype("int"), rf.predict(X_test))
print(AUC)
```

```python
print(classification_report(y_test.astype("int"), rf.predict(X_test)))
```

```python
cm = metrics.confusion_matrix(y_test.astype("int"), rf.predict(X_test))

plt.figure(figsize=(9, 9))
sns.heatmap(cm, annot=True, fmt=".3f", linewidths=0.5, square=True, cmap="Blues_r")
plt.ylabel("Actual label")
plt.xlabel("Predicted label")
plt.title("Random Forest", size=25)

# We have 249 Correct Predictions for Conversions, 124 False Positives, and 19 False Negatives
```

```python
tree = rf.estimators_[0]
# Import tools needed for visualization

# Pull out one tree from the forest
tree = rf.estimators_[5]
# Export the image to a dot file
export_graphviz(
    tree, out_file="tree.dot", feature_names=feature_list, rounded=True, precision=1
)
# Use dot file to create a graph
(graph,) = pydot.graph_from_dot_file("tree.dot")
# Write graph to a png file
graph.write_png("tree.png")

from IPython.display import Image

Image(filename="tree.png")
```

```python
# Get numerical feature importances
importances = list(rf.feature_importances_)
# List of tuples with variable and importance
feature_importances = [
    (feature, round(importance, 2))
    for feature, importance in zip(feature_list, importances)
]
# Sort the feature importances by most important first
feature_importances = sorted(feature_importances, key=lambda x: x[1], reverse=True)
# Print out the feature and importances
[print("Variable: {:20} Importance: {}".format(*pair)) for pair in feature_importances]
```

 &nbsp;&nbsp; 

<center><font size="10">Randomized Search</font></center>

```python
# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start=200, stop=2000, num=10)]
# Number of features to consider at every split
max_features = ["auto", "sqrt"]
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(10, 110, num=11)]
max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 4]
# Method of selecting samples for training each tree
bootstrap = [True, False]

random_grid = {
    "n_estimators": n_estimators,
    "max_features": max_features,
    "max_depth": max_depth,
    "min_samples_split": min_samples_split,
    "min_samples_leaf": min_samples_leaf,
    "bootstrap": bootstrap,
}

random_forest = RandomForestClassifier()
rf_random = RandomizedSearchCV(
    estimator=random_forest,
    param_distributions=random_grid,
    n_iter=100,
    cv=3,
    verbose=2,
    random_state=42,
    n_jobs=-1,
)
rf2 = rf_random.fit(X_train, y_train.astype("int"))


AUC2 = roc_auc_score(y_test.astype("int"), rf2.predict(X_test))
print(AUC2)
```

```python
rf2.best_params_
```

```python
print(classification_report(y_test.astype("int"), rf2.predict(X_test)))
```

```python
cm = metrics.confusion_matrix(y_test.astype("int"), rf2.predict(X_test))

plt.figure(figsize=(9, 9))
sns.heatmap(cm, annot=True, fmt=".3f", linewidths=0.5, square=True, cmap="Blues_r")
plt.ylabel("Actual label")
plt.xlabel("Predicted label")
plt.title("Random Forest", size=25)

# We have 190 Correct Predictions for Conversions, 6 False Positives, and 78 False Negatives
```

```python
label_encoder = LabelEncoder()
y_train3 = label_encoder.fit_transform(y_train)
y_test3 = label_encoder.fit_transform(y_test)


model_GBC = GradientBoostingClassifier()
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
n_scores = cross_val_score(
    model_GBC,
    X_train,
    y_train3,
    scoring="accuracy",
    cv=cv,
    n_jobs=-1,
    error_score="raise",
)

gb3 = model_GBC.fit(X_train, y_train3)
AUC3 = roc_auc_score(y_train.astype("int"), gb3.predict(X_train))
print(AUC3)
```

```python
print(classification_report(y_test.astype("int"), gb3.predict(X_test)))
```

```python
cm = metrics.confusion_matrix(y_test.astype("int"), gb3.predict(X_test))

plt.figure(figsize=(9, 9))
sns.heatmap(cm, annot=True, fmt=".3f", linewidths=0.5, square=True, cmap="Blues_r")
plt.ylabel("Actual label")
plt.xlabel("Predicted label")
plt.title("Random Forest", size=25)

# We have 202 Correct Predictions for Conversions, 6 False Positives, and 66 False Negatives
```
